```python
#import pandas to read excel
import pandas as pd

#reads the CSV file into a data table
dt = pd.read_csv("CaliforniaOffices.csv")
dt.head()
```

```python
#imports folium interactive map library
import folium

#sets the center location of the map
loc_center = [dt['Latitude'].mean(), dt['Longitude'].mean()]
```

```python
#Creates the map
map1 = folium.Map(location = loc_center, tiles='Openstreetmap', zoom_start = 5, control_scale=True)

#Loops through data table and adds marker to the map
for index, loc in dt.iterrows():
    folium.CircleMarker([loc['Latitude'], loc['Longitude']], radius=2, weight=5, popup=loc['CSO']).add_to(map1)
folium.LayerControl().add_to(map1)

#Displays the map
map1
```

```python
#Saves the map as an interactive html
map1.save('Offices.html')
```
