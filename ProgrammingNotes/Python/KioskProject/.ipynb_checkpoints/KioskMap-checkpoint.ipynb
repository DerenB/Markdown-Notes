```python
#import pandas to read excel
import pandas as pd

#read in CSV file data into data table
dt = pd.read_csv("KioskLocationData.csv")
dt.head()
```

```python
#import folium library
import folium

#set the map center location
loc_center = [dt['Latitude'].mean(), dt['Longitude'].mean()]

#Clear previous Map
map1 = folium.Map()

#Create the Map
map1 = folium.Map(location = loc_center, tiles='Openstreetmap', zoom_start = 9, control_scale=True)

#Display Blank Map
map1
```

```python
#Loop to add data points to map
for i in range(0,len(dt)):
    folium.CircleMarker(
        location=[dt.iloc[i]['Latitude'], dt.iloc[i]['Longitude']],
        popup=dt.iloc[i]['DBA'],
        radius=float(4),
        weight=float(2),
        color='#000000',
        opacity = 1,
        fill=True,
        fill_color='#004998',
        fillOpacity = 1
    ).add_to(map1)
    
#Display updated Map
map1
```

```python
#Save the map as an interactive HTML
map1.save('KioskLocations.html')
```
